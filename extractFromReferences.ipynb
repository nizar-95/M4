# Done:
- from dataframe of the df_references we convert to txt
- set every citation in new line in new dataframe
- extract from every citation the year, authors..
- set manually the Target Variable "Relevance of Reference"

```python
import pandas as pd
import re

# Open the text file and read the contents into a string variable
with open('data.txt', 'r') as file:
    input_str = file.read()

# Split the input string into individual lines
lines = input_str.split('\n')

# Create empty lists to store the extracted citations, years, and author names
citations = []
years = []
authors = []
titles = []

# Iterate over the lines and extract any citations, years, and author names
for line in lines:
    # Check if the line contains a citation using a regular expression
    if re.search('\(\d{4}\)\.?\s', line):
        # Extract the citation using a regular expression
        citation = re.sub('\s*\(\d{4}\)\.?\s*', '', line)
        # Extract the year using a regular expression
        year = re.search('\(\d{4}\)', line).group()
        year = year.strip('()')
        # Extract the author names using a regular expression
        authors_str = re.search('^(.*)\(\d{4}\)', line).group(1)
        authors_str = re.sub('\.$', '', authors_str)
        authors_list = authors_str.split(', ')
        # title = authors_list[1]
        # Add the citation, year, and author names to their respective lists
        citations.append(citation)
        years.append(year)
        authors.append(authors_list)
        # titles.append(title)

# Create a new DataFrame with 'Year', 'Citations', and 'Authors' columns containing the extracted data
df = pd.DataFrame({'Year': years, 'Citations': citations, 'Authors': authors})

# Remove any leading or trailing whitespace from the citations and author names
df['Citations'] = df['Citations'].str.strip()
df['Authors'] = df['Authors'].apply(lambda x: [author.strip() for author in x])
# df['Titles'] = df['Titles'].str.strip()

# Remove any empty citations (if any)
df = df[df['Citations'] != '']

# Reset the index of the DataFrame
df = df.reset_index(drop=True)

# Display the resulting DataFrame
display(df)
```

```python
# Create a new column called 'Relevance'
df['Relevance'] = 0

# Set the value of 'Relevance' to 1 for the specified lines
df.loc[[14, 15, 20, 44, 45], 'Relevance'] = 1
display(df)
```
